```python
import torch
from transformers import pipeline

pipe1 = pipeline("question-answering", model="lserinol/bert-turkish-question-answering")
```

```python
pipe2 = pipeline("question-answering", model="savasy/bert-base-turkish-squad")
```

```python
pipe3 = pipeline("question-answering", model="Izzet/qa_ytu_bert-base-turkish")
```

```python
pipe4 = pipeline("question-answering", model="Aybars/XLM_Turkish")
```

```python
pipe5 = pipeline("question-answering", model="husnu/bert-base-turkish-128k-cased-finetuned_lr-2e-05_epochs-3")
```

```python
qa_pipelines = [pipe1, pipe2, pipe3, pipe4, pipe5]
```

```python
context = """İtalyanın en meşhur yemekleri arasında pizza ve makarna bulunmaktadır.
     Pizza, İtalyan mutfağının en ikonik lezzetlerinden biridir.
     Çeşitli malzemelerle pişirilen ince hamurlu bir yemektir.
     Makarna da İtalyan mutfağının temel bir parçasıdır.
     Farklı soslarla sunulan çeşitli makarna çeşitleri bulunmaktadır.
     Pizza, çeşitli malzemelerle pişirilen ince hamurlu bir yemektir.
     Makarna ise farklı soslarla sunulan çeşitli makarna çeşitlerinden oluşur.
     Pizza ve makarna, genellikle İtalyan mutfağına özgü malzemelerle hazırlanır.
     Pizza ve makarna, İtalyan mutfağına özgü lezzetler olarak dünyaya yayılmıştır.
     İtalyan mutfağı, dünya çapında popülerlik kazanmış ve birçok ülkede sevilen bir mutfak kültürü haline gelmiştir.
     Makarna sosları, genellikle çeşitli sebzeler, et veya peynir gibi malzemelerden yapılır.
     İtalyan mutfağında kullanılan soslar genellikle doğal ve taze malzemelerle hazırlanır. İtalyan mutfağında genellikle şarap içilir. Ancak, diğer içecekler de tercih edilebilir.
     Özellikle yemeklerle uyumlu şarap çeşitleri, İtalyan mutfağının önemli bir parçasını oluşturur."""
```

```python
questions = ["İtalyanın en meşhur yemekleri nelerdir?",
             "Pizza hangi malzemelerden yapılır?",
             "Hangi olayla dünyaya yayılmıştır",
             "Soslar hangi malzemelerden yapılır?",
             "İçecek olarak ne tercih ederler?"]
```

```python
models = ["lserinol/bert-turkish-question-answering",
           "savasy/bert-base-turkish-squad",
          "Izzet/qa_ytu_bert-base-turkish",
          "Aybars/XLM_Turkish",
          "husnu/bert-base-turkish-128k-cased-finetuned_lr-2e-05_epochs-3"]
```

```python
for question in questions:
    print(question)
    for pipe, model in zip(qa_pipelines, models):
        result = pipe(question=question, context=context)
        print("Model: " + model)
        print(f"Answer: {result['answer']}")
    print()
```
